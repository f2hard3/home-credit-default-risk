```python
import pandas as pd

cust_dict = {
    'customer_name':['Alice', 'Tom', 'James', 'Yerin', 'Min', 'Katherine', 'John', 'Park', 'Maria', 'Derik', 'Jin'],
    'cust_country':['US', 'GB', 'US', 'KOR', 'KOR', 'GB', 'US', 'KOR', 'US', 'GB', 'KOR'],
    'gender':['F', 'M', 'M', 'F', 'M', 'F', 'M', 'M', 'F', 'M', 'F'],
    'grade':[3, 1, 2, 5, 4, 3, 5, 7, 2, 4, 5],
    'age':[25, 34, 26, 33, 67, 29, 54, 21, 77, 29, 16]
}
customer = pd.DataFrame(cust_dict)

customer.head(11)
```

```python
male_customer = customer[customer['gender'] == 'M']
female_customer = customer[customer['gender'] == 'F']

print(male_customer)
print()
print(female_customer)
```

```python
country_agg = customer.groupby('cust_country')['grade'].sum()
country_agg = country_agg.reset_index()
country_agg.head()
```

```python
country_male_grade_sum = male_customer.groupby('cust_country')['grade'].sum()
country_female_grade_sum = female_customer.groupby('cust_country')['grade'].sum()

country_male_grade_sum = country_male_grade_sum.reset_index()
country_female_grade_sum = country_female_grade_sum.reset_index()

print(country_male_grade_sum)
print(country_female_grade_sum)
```

```python
country_agg = country_agg.merge(country_male_grade_sum, on='cust_country', how='left')
country_agg = country_agg.merge(country_female_grade_sum, on='cust_country', how='left')

country_agg
```

```python
country_agg.columns = ['cust_country', 'sum_grade', 'sum_grade_male', 'sum_grade_female']

country_agg
```
