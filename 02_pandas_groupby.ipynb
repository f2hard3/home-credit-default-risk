```python
import pandas as pd

cust_dict = {
    'customer_name':['Alice', 'Tom', 'James', 'Yerin', 'Min', 'Katherine', 'John', 'Park', 'Maria', 'Derik', 'Jin'],
    'cust_country':['US', 'GB', 'US', 'KOR', 'KOR', 'GB', 'US', 'KOR', 'US', 'GB', 'KOR'],
    'gender':['F', 'M', 'M', 'F', 'M', 'F', 'M', 'M', 'F', 'M', 'F'],
    'grade':[3, 1, 2, 5, 4, 3, 5, 7, 2, 4, 5],
    'age':[25, 34, 26, 33, 67, 29, 54, 21, 77, 29, 16]
}
customer = pd.DataFrame(cust_dict)

customer.head(11)
```

```python
country_group = customer.groupby('cust_country')
type(country_group)
```

```python
country_group.head()
```

```python
country_agg = pd.DataFrame()

country_agg['sum_grade'] = country_group['grade'].sum()
country_agg['max_grade'] = country_group['grade'].max()
country_agg.head()
```

```python
country_agg['avg_age'] = country_group['age'].mean()
country_agg['max_age'] = country_group['age'].max()
country_agg.head()
```

```python
print(country_agg.index)
country_agg = country_agg.reset_index()
country_agg.head()
```

```python
country_agg1 = country_group['grade'].agg(['sum', 'max'])
country_agg2 = country_group['age'].agg(['mean', 'max'])

print(country_agg1.head())
print(country_agg2.head())
```

```python
country_agg = country_agg1.merge(country_agg2, on='cust_country', how='left')
print(country_agg)
```

```python
country_agg = pd.DataFrame()

country_agg[['sum_grade', 'max_grade']] = country_group['grade'].agg(['sum', 'max'])
country_agg[['mean_age', 'max_age']] = country_group['age'].agg(['mean', 'max'])

country_agg.head()
```

```python
agg_dict = {
  'grade': ['sum', 'max'],
  'age': ['mean', 'max']
}

country_agg = country_group.agg(agg_dict)
country_agg.head()
```
